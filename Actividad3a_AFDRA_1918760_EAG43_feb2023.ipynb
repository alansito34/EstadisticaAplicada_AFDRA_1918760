**Ventas es una base de datos que contiene las ventas mensuales de una tienda X para los años 2015, 2016 y 2017**
1. Crea una nueva columna que se llame Ganancias y que contenga la diferencia entre ventas-gastos

```python
ventas["Ganancias"]=ventas["Ventas"]-ventas["Gastos"]
ventas
```

3. Calcula e imprime las ventas totales, gastos totales y ganacias totales de todo el periodo
Ejemplo de mensaje= las ventas totales son:
etc

```python
Ventas=ventas["Ventas"].sum()
Gastos=ventas["Gastos"].sum()
Ganancias=ventas["Ganancias"].sum()
print("Las ventas totales son:", Ventas)
print("Las gastos totales son:", Gastos)
print("Las ganancias totales son", Ganancias)
```

4. Calcula e imprime las ventas promedio, gastos promedio y ganacias promedio del periodo
Ejemplo de mensaje= las ventas promedio son:
etc

```python
Ventas=ventas["Ventas"].mean()
Gastos=ventas["Gastos"].mean()
Ganancias=ventas["Ganancias"].mean()
print("Las ventas promedio son:", Ventas)
print("Las gastos promedio son:", Gastos)
print("Las ganancias promedio son:", Ganancias)
```

5. Calcula e imprime las ventas promedio de 2015

```python
a2015=ventas[ventas["Año"] == 2015]
prom2015=a2015["Ventas"].mean()
print("La venta promedio en el año 2015:", prom2015)
```

6. Calcula e imprime las ventas totales para los 3 meses de enero. Es decir la suma de enero de 2015, enero de 2016 y enero de 2017

```python
a2015=ventas[ventas["Mes"] == "Enero"]
prom2015=a2015["Ventas"].sum()
print("Las ventas totales para los 3 mes de enero de los años 2015, 2016 y 2017 es:", prom2015)
```

7. Calcula las ventas promedio por mes, guardalo en un dataframe y por ultimo muestra ordenado por ventas promedio de menor a mayor

```python
prom_mes=ventas.groupby("Mes").mean()["Ventas"]
promedio=pd.DataFrame(prom_mes)
promedio=promedio.sort_values("Ventas",ascending=True)
promedio.rename(columns={"Ventas":"Ventas promedio"},inplace=True)
promedio
```
