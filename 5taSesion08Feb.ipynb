```python
mun_nl=pd.DataFrame({"Municipio":["Monterrey","Guadalupe","Apodaca","San Nicolas","Santiago"],"Nacimientos":[500,630,768,400,250]})
mun_nl
```

```python
plt.bar(mun_nl["Municipio"],mun_nl["Nacimientos"], color="black")
plt.show
```

```python
mun_nl=mun_nl.sort_values("Nacimientos",ascending=False)
mun_nl
plt.bar(mun_nl["Municipio"],mun_nl["Nacimientos"], color="black")
```

```python
plt.bar(mun_nl["Municipio"],mun_nl["Nacimientos"],color=["blue","green","red","yellow","gray"])
plt.title("Nacimientos por municipio")
plt.xlabel("Municipios de N.L.")
plt.ylabel("Cantidad de nacimientos")
plt.yticks([0,200,400,600,800])
plt.show()
```

```python
fig,(panel1,panel2)=plt.subplots(2,1) # subgraficas, su nombre
fig.tight_layout()
x1=1 #eje x de grafico 1
y1=1 #eje y de grafico 1
panel1.plot(x,y,label="nombre de linea")
panel1.title.set_text("Grafica 1")

x2=mun_nl["Municipio"]
y2=mun_nl["Nacimientos"]
panel2.bar(x2,y2)
panel2.title.set_text("Grafica 2")
```

```python
fig,(panel1,panel2)=plt.subplots(2,1) # subgraficas, su nombre
fig.tight_layout()
x1=1 #eje x de grafico 1
y1=1 #eje y de grafico 1
panel1.scatter(x,y,label="nombre de linea")
panel1.title.set_text("Grafica 1")

x2=mun_nl["Municipio"]
y2=mun_nl["Nacimientos"]
panel2.barh(x2,y2)
panel2.title.set_text("Grafica 2")
```

```python
plt.barh(mun_nl["Municipio"],mun_nl["Nacimientos"])
plt.ylabel("Municipios de N.L.")
plt.xlabel("Cantidad de nacimientos")
plt.title("Natalidad")
```
