6. Elige seis variables para ajustar un modelo de cox de riesgos porporcionales. Antes de correr el modelo explica en qu√© sentido y por qu√© afectan el riesgo de pagar el pr√©stamo cada una de las variables elegidas


```python
from lifelines import CoxPHFitter
```

**1. credit_history:** Para conocer que tan riesgoso sera otorgar un credito o no a un tercero, estrictamente se requiere conocer su historial crediticio. Variable dummy = 'critical_account'.

**2. amount:** La cantidad de prestamo es inportante, ya que puede que si es muy grande, no se pueda tener de vuelta la cantidad de dinero prestada.

**3. purpose:** Conocer el motivo del prestamo nos da una idea mas certera del proposito de otorgar dicho prestamo. Variable dummy= 'education'.

**4. age:** La edad de alguien no adulto, puede ser financieramente inestable.

**5. savings_account_status:** Al tener conocimiento sobre los fondos de ahorro del prestatario, nos hara saber como es el comportamiento financiero del mencionado. Variable dummy= 'below_100'.

**6. property:** Contar con bienes, concede un aseguramiento economico mayor.. Variable dummy= 'no_property'.  

```python
ch = base[['duration', 'credit_history', 'full_repaid']]
ch['C.A.'] = np.where(base['credit_history'] == 'critical_account', 1, 0)
ch = ch.drop(['credit_history'], axis = 1)

cox_ch = CoxPHFitter().fit(ch, 'duration', 'full_repaid')
cox_ch.print_summary()
```

```python
amount = base[['duration', 'amount', 'full_repaid']]

cox_amount = CoxPHFitter().fit(amount, 'duration', 'full_repaid')
cox_amount.print_summary()
```

```python
purpose=base[['duration', 'purpose', 'full_repaid']]
purpose['Education']=np.where(base['purpose']=='education', 1, 0)
purpose=purpose.drop(['purpose'], axis=1)

cox_purpose=CoxPHFitter().fit(purpose, 'duration', 'full_repaid')
cox_purpose.print_summary()
```

```python
age = base[['duration', 'age', 'full_repaid']]

cox_age = CoxPHFitter().fit(age, 'duration', 'full_repaid')
cox_age.print_summary()
```

```python
sa = base[['duration', 'savings_account_status', 'full_repaid']]
sa['S.A.'] = np.where(base['savings_account_status'] == 'below_100', 1, 0)
sa = sa.drop(['savings_account_status'], axis = 1)

cox_sa = CoxPHFitter().fit(sa, 'duration', 'full_repaid')
cox_sa.print_summary()
```

```python
property = base[['duration', 'full_repaid', 'property']]
property['N.P.'] = np.where(base['property'] == 'no_property', 1, 0)
property=property.drop(['property'], axis=1) 

cox_property = CoxPHFitter().fit(property, 'duration', 'full_repaid')
cox_property.print_summary()
```
