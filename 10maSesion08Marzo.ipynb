**Objetivo** es medir el riesgp de volver a ser arrestado en funcion de las variables que tenemos

**hipotesis**

1. Si reciben dinero (fin) al salir de la carcel disminuye el riesgo de volver a ser arrestado porque no se tiene la necesidad de robar.

2. Los jovenes son mas inexpertos y los atrapan mas facil, por tanto tienen mas riesgo de volver a ser arrestado.

3. Los negros (1970) tienen mas riesgo de volver a ser arrestado.

4. No influyen si tenias trabajo en el riesgo de volver a ser arrestado.

5. b) No robas porque si te arrestan se qudan sin el sustento del hogar.

6. Si tienes libertad condicional el riesgo de volver a ser arrestado disminuye.

7. Entre mas veces hayas sido arrestado aumenta el riesgo de volver a ser arrestado.

```python
# Modelo de Cox de Riesgos Proporcionales
cph=CoxPHFitter().fit(rossi,"week","arrest")
cph.print_summary()
```

El signo de "coef" nos dice si aumenta/disminuye el riesgo

El exp(coef) te dice el efecto. Si coef<0   1 - exp(coef)

                        Si coef>0   exp(coef) - 1

Suponer "fin" tiene efecto. 
Recibir dinero disminuye el riesgo de volver a ser arrestado. 

1 - 0.68 = 0.32

Los que reciben dinero tienen 32% ,emps riesgo de ser arrestados que alguien que no recibe ayuda.

La edad disminuye el riesgo de volver a ser arrestado

1 - 0.94 = 0.06 => 6%

En promedio por cada año cumplido el riesgo de ser arrestado disminuye en 6%.

Tener arrestos anteriores aumenta el riesgo de volver a ser arrestado

En promedio por cada arresto anterior, el riesgo de ser arrestado aumenta 10%

```python
cph.plot_partial_effects_on_outcome(covariates="age", values=[20,30,40,50])
```
